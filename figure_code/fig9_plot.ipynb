```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


def get_only_acc(filename) : # log에서 accuracy만 추출

    with open(filename) as f :
        text = f.read()

    res = []
    for t in text.split('Accuracy: ')[1:] :
        res.append(float(t[:5]))

    return res


def get_only_f1(filename) : # log에서 F1 score만 추출

    with open(filename) as f :
        text = f.read()

    f1 = [0 for _ in range(10)]

    res = []
    for t in text.split('f-score: ')[1:] :
        res.append(float(t[:5]))
    f1 = f1 + res
    return f1


def get_list_mean(list1, list2, num) : # 두개 평균 내기
    temp_list = []
    for one, two in zip(list1, list2) :
        temp_list.append((one+two)/num)
    return temp_list

res_k21 = get_only_acc('log/base/size21.txt')
res_k31 = get_only_acc('log/base/size31.txt')
res_k41 = get_only_acc('log/base/size41.txt')
res_k51 = get_only_acc('log/base/size51.txt')
res_k22 = get_only_acc('log/base/size22.txt')
res_k32 = get_only_acc('log/base/size32.txt')
res_k42 = get_only_acc('log/base/size42.txt')
res_k52 = get_only_acc('log/base/size52.txt')

res_k2 = get_list_mean(res_k21, res_k22, 2)
res_k3 = get_list_mean(res_k31, res_k32, 2)
res_k4 = get_list_mean(res_k41, res_k42, 2)
res_k5 = get_list_mean(res_k51, res_k52, 2)

# plot
plt.figure(figsize=(5,4))
plt.xlim(10,75)
plt.ylim(70,100)
plt.plot(res_k2, color='#FF3D3D', label='T=2')
plt.plot(res_k3, color='#3A9EFF', label='T=3')
plt.plot(res_k4, color='#B77CED', label='T=4')
plt.plot(res_k5, color='#209090', label='T=5')
plt.legend()
plt.xlabel('Epoch')
plt.ylabel('Test Acc.(%)')
plt.title('Inst 40%')
plt.show()
```

```python
res_k21 = get_only_f1('log/base/size21.txt')
res_k31 = get_only_f1('log/base/size31.txt')
res_k41 = get_only_f1('log/base/size41.txt')
res_k51 = get_only_f1('log/base/size51.txt')
res_k22 = get_only_f1('log/base/size22.txt')
res_k32 = get_only_f1('log/base/size32.txt')
res_k42 = get_only_f1('log/base/size42.txt')
res_k52 = get_only_f1('log/base/size52.txt')

res_k2 = get_list_mean(res_k21, res_k22, 2)
res_k3 = get_list_mean(res_k31, res_k32, 2)
res_k4 = get_list_mean(res_k41, res_k42, 2)
res_k5 = get_list_mean(res_k51, res_k52, 2)

plt.figure(figsize=(5,4))
plt.xlim(10,75)
plt.ylim(0.86,1)
plt.plot(res_k2, '-.', color='#FF3D3D', label='T=2')
plt.plot(res_k3, '-.', color='#3A9EFF', label='T=3')
plt.plot(res_k4, '-.', color='#B77CED', label='T=4')
plt.plot(res_k5, '-.', color='#209090', label='T=5')
plt.legend()
plt.xlabel('Epoch')
plt.ylabel('F-score')
plt.title('Inst 40%')
plt.show()
```
